e) Ma tráº­n nghá»ch Äáº£o



```python
def invert_matrix(matrix):
    result = np.linalg.inv(matrix)
    return result

matrix = np.array([[-2,6],[8,-4]])
print(invert_matrix(matrix))
```

```python
def invert_matrix(matrix):
    detA = matrix[0,0]*matrix[1,1]-matrix[0,1]*matrix[1,0]
    matrix_new = np.zeros((2,2), dtype=int)
    matrix_new[0,0] = matrix[1,1]
    matrix_new[1,1] = matrix[0,0]
    matrix_new[0,1] = -matrix[0,1]
    matrix_new[1,0] = -matrix[1,0]
    if detA != 0:
        return (1/detA) * matrix_new
    return None

matrix = np.array([[-2,6],[8,-4]])
print(invert_matrix(matrix))
```

f) Eigenvectors vÃ  eigenvalues



```python
def compute_eigenvectors_eigenvalues(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    return eigenvalues, eigenvectors

maxtrix = np.array([[0.9, 0.2],[0.1, 0.8]])
value, vector = compute_eigenvectors_eigenvalues(maxtrix)
print(value)
print(vector)
```

g) Cosine Similarity



```python
def compute_cosine_similarity(vector1, vector2):
    xy = vector1 @ vector2
    x1 = np.linalg.norm(vector1)
    x2 = np.linalg.norm(vector2)
    return xy/(x1*x2)

vector1 = np.array([1,2,3])
vector2 = np.array([18,20,30])
print(compute_cosine_similarity(vector1, vector2))
```
