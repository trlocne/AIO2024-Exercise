```python
# Exercise 1:
def sibling_window():
    num_list = [3, 4, 5, -44, 5, 10, 12, 33, 1]
    k = 3
    output = []
    sibling = []
    
    if len(num_list) < k:
        output.append(max(num_list))
        return output
    
    for item in num_list:
        if k > 0:
            sibling.append(item)
            k-=1
        else:
            output.append(max(sibling))
            sibling.pop(0)
            sibling.append(item)
    
    output.append(max(sibling))
    return output
    
print(sibling_window())
```

```python

# Exercise 2: Count chars
def count_chars(s):
    dict = {x: 0 for x in s}
    for c in s:
        dict[c]+=1
    return dict
    
count_chars("Happiness")
```

```python
def word_counts(file_path):
    with open(file_path, 'r') as f:
        arr_word = f.read().lower().split()
        dict = {x: 0 for x in arr_word}
        for c in arr_word:
            dict[c]+=1
    
    return dict

print(word_counts("P1_data.txt"))
```

```python
def levenshtein(str1, str2):
    m = len(str1)
    n = len(str2)
    levent = [[0]*(n+1) for _ in range(m+1)]
    
    for i in range(1, m + 1):
        levent[i][0] = i
        
    for j in range(1, n + 1):
        levent[0][j] = j
        
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if str1[i-1] == str2[j - 1]:
               levent[i][j] = levent[i-1][j-1]
            else:
               levent[i][j] = min(levent[i-1][j-1], levent[i-1][j], levent[i][j-1]) + 1
               
    return levent[m][n]
```
