```python
import torch
import torch.nn as nn
```

```python
class Softmax(nn.Module):
    def __init__(self):
        super().__init__()
    
    def forward(self, data):
        data = torch.exp(data)
        sum_data = torch.sum(data)
        return data / sum_data
        
```

```python
data = torch . Tensor([1 , 2 , 300000])
softmax = Softmax()
output = softmax(data)
output
```

```python
class softmax_stable(nn.Module):
    def __init__(self):
        super().__init__()
    
    def forward(self, data):
        c = torch.max(data)
        data = torch.exp(data - c)
        sum_data = torch.sum(data)
        return data / sum_data
        
```

```python
data = torch.Tensor([1 , 2 , 3])
softmax_stable = softmax_stable()
output = softmax_stable(data)
output
```

```python
class Person:
    def __init__(self, name: str, yob: int):
        self.name = name
        self.yob = yob
    
    def getAge(self):
        return self.yob
        
class Ward:
    def __init__(self, name: str):
        self.name = name
        self.listPeople = list()
        
    def addPerson(self, person: Person):
        self.listPeople.append(person)
        
    def describe(self):
        print(f"Ward Name: {self.name}")
        for person in self.listPeople:
            person.describe()
            
    def count_doctor(self):
        cnt = 0
        for person in self.listPeople:
            if type(person) is Doctor:
                cnt += 1
        return cnt
    
    def sort_age(self):
        self.listPeople = sorted(self.listPeople, key = lambda x : x.getAge(), reverse = True)
        
    def compute_average(self):
        sum = 0
        i = 0
        for person in self.listPeople:
            if type(person) is Teacher:
                sum += person.getAge()
                i += 1
            
        return sum // i

class Student(Person):
    def __init__(self, name: str, yob: int, grade: str):
        super().__init__(name = name, yob = yob)
        self.grade = grade
        
    def describe(self):
        print(f"Student - Name: {self.name} - YoB: {self.yob} - Grade:{self.grade}")

class Teacher(Person):
    def __init__(self, name: str, yob: int, subject: str):
        super().__init__(name = name, yob = yob)
        self.subject = subject
        
    def describe(self):
        print(f"Teacher - Name: {self.name} - YoB: {self.yob} - Grade:{self.subject}")

class Doctor(Person):
    def __init__(self, name: str, yob: int, specialist: str):
        super().__init__(name = name, yob = yob)
        self.specialist = specialist
        
    def describe(self):
        print(f"Doctor - Name: {self.name} - YoB: {self.yob} - Grade:{self.specialist}")
```

```python
class MyStack:
    def __init__(self, cap: int):
        self.cap = cap
        self.cnt = 0
        self.arr = list()

    def is_empty(self):
        return self.cnt == 0
    
    def is_full(self):
        return self.cnt >= self.cap
    
    def pop(self):
        if self.cnt == 0:
            return None
        
        self.cnt -= 1
        val = self.arr[0]
        del self.arr[0]
        return val
    
    def push(self, value):
        if self.is_full():
            return False
        
        self.cnt += 1
        self.arr.insert(0, value)
        return True
    
    def top(self):
        if self.cnt == 0:
            return None
        
        return self.arr[0]      
        
```

```python
stack1 = MyStack(cap = 5)
stack1.push(1)
stack1.push(2)
stack1.push(3)
stack1.push(4)
stack1.push(5)
stack1.pop()
stack1.top()
```

```python
class MyQueue:
    def __init__(self, cap: int):
        self.cap = cap
        self.cnt = 0
        self.arr = list()

    def is_empty(self):
        return self.cnt == 0
    
    def is_full(self):
        return self.cnt >= self.cap
    
    def dequeue(self):
        if self.cnt == 0:
            return None
        
        self.cnt -= 1
        val = self.arr[0]
        del self.arr[0]
        return val
    
    def enqueue(self, value):
        if self.is_full():
            return False
        
        self.cnt += 1
        self.arr.append(value)
        return True
    
    def front(self):
        if self.cnt == 0:
            return None
        
        return self.arr[0]   
```
